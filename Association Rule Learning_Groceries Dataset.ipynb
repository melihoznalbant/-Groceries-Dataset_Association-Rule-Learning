```python
# Creating a DataFrame with a new level of column labels whose innermost level consists of summarized index labels.

def create_invoice_product_df (dataframe): 
    return dataframe.groupby(['Member_number', "itemDescription"])['itemDescription'].count().unstack().fillna(0).applymap(lambda x: 1 if x > 0 else 0)

  
inv_pro_df = create_invoice_product_df(df)
```

```python
inv_pro_df
```

```python
# Calculating possible product pairs according to a support threshold value to be determined and creating the final final table by making eliminations according 
# to the determined support value in each iteration.

frequent_itemsets = apriori (inv_pro_df, min_support=0.05, use_colnames=True)
```

```python
frequent_itemsets.sort_values("support", ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="support",min_threshold=0.01)
```

```python
rules.head()
```
